```python
import pandas as pd
import numpy as np
```

```python
#read csv files with pandas dataframe - testing janurady 2019 df
jan_2019 = pd.read_csv('all_data/JC-201901-citibike-tripdata.csv')
jan_2019
```

```python
#read all 2019 & 2020 data
feb_2019 = pd.read_csv('all_data/JC-201902-citibike-tripdata.csv')
mar_2019 = pd.read_csv('all_data/JC-201903-citibike-tripdata.csv')
apr_2019 = pd.read_csv('all_data/JC-201904-citibike-tripdata.csv')
may_2019 = pd.read_csv('all_data/JC-201905-citibike-tripdata.csv')
jun_2019 = pd.read_csv('all_data/JC-201906-citibike-tripdata.csv')
jul_2019 = pd.read_csv('all_data/JC-201907-citibike-tripdata.csv')
aug_2019 = pd.read_csv('all_data/JC-201908-citibike-tripdata.csv')
sep_2019 = pd.read_csv('all_data/JC-201909-citibike-tripdata.csv')
oct_2019 = pd.read_csv('all_data/JC-201910-citibike-tripdata.csv')
nov_2019 = pd.read_csv('all_data/JC-201911-citibike-tripdata.csv')
dec_2019 = pd.read_csv('all_data/JC-201912-citibike-tripdata.csv')
jan_2020 = pd.read_csv('all_data/JC-202001-citibike-tripdata.csv')
feb_2020 = pd.read_csv('all_data/JC-202002-citibike-tripdata.csv')
mar_2020 = pd.read_csv('all_data/JC-202003-citibike-tripdata.csv')
apr_2020 = pd.read_csv('all_data/JC-202004-citibike-tripdata.csv')
may_2020 = pd.read_csv('all_data/JC-202005-citibike-tripdata.csv')
jun_2020 = pd.read_csv('all_data/JC-202006-citibike-tripdata.csv')
jul_2020 = pd.read_csv('all_data/JC-202007-citibike-tripdata.csv')
aug_2020 = pd.read_csv('all_data/JC-202008-citibike-tripdata.csv')
sep_2020 = pd.read_csv('all_data/JC-202009-citibike-tripdata.csv')
oct_2020 = pd.read_csv('all_data/JC-202010-citibike-tripdata.csv')
nov_2020 = pd.read_csv('all_data/JC-202011-citibike-tripdata.csv')
dec_2020 = pd.read_csv('all_data/JC-202012-citibike-tripdata.csv')
```

```python
#combine all 2019 data
citibike_2019 = pd.concat([jan_2019,feb_2019,mar_2019, apr_2019, may_2019, jun_2019, jul_2019, aug_2019, sep_2019, oct_2019, nov_2019, dec_2019])
citibike_2019
```

```python
#combine all 2020 data
citibike_2020 = pd.concat([jan_2020, feb_2020, mar_2020, apr_2020, may_2020, jun_2020, jul_2020, aug_2020, sep_2020, oct_2020, nov_2020, dec_2020])
citibike_2020
```

```python
#combine 2019 and 2020 data
citibike_all = pd.concat([citibike_2019, citibike_2020])
citibike_all
```
